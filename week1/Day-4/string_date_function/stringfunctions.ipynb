```sql
--QUESTION 2: Order Delivery Delay Analysis

select upper(customer_name),datediff(order_date,coalesce(delivery_date,curdate())) as  daliverys_days ,coalesce(delivery_date,curdate()),FLOOR(order_amount * 10) / 10,
 CASE 
        WHEN delivery_date IS NULL THEN 'Pending'
        WHEN DATEDIFF(delivery_date, order_date) = 0 THEN 'Same-day'
        WHEN DATEDIFF(delivery_date, order_date) > 3 THEN 'Delayed'
        ELSE 'On-time'
    END AS delivery_status

FROM orders_delivery;
```

```sql
--QUESTION 5: Loan EMI Risk Categorization

SELECT 
    loan_id,

    -- Uppercase customer name
    UPPER(customer_name) AS customer_name,

    loan_amount,
    interest_rate,
    loan_start,

    -- 1. Years since loan start
    TIMESTAMPDIFF(YEAR, loan_start, CURRENT_DATE) AS years_since_start,

    -- 2. Monthly interest rate using power logic
    POWER(1 + (interest_rate/100), 1/12) - 1 AS monthly_interest_rate,

    -- 3. EMI Calculation (Rounded)
    ROUND(
        loan_amount * 
        ( (POWER(1 + (interest_rate/100)/12, 12 * 5) * (interest_rate/100)/12) 
        / (POWER(1 + (interest_rate/100)/12, 12 * 5) - 1) )
    , 0) AS emi,

    -- 4. Risk Category
    CASE 
        WHEN interest_rate > 9 THEN 'High Risk'
        WHEN interest_rate BETWEEN 8 AND 9 THEN 'Medium Risk'
        ELSE 'Low Risk'
    END AS risk_category

FROM loan_details;
```

```sql
CREATE TABLE insurance_policies (

policy_id INT,

holder_name VARCHAR(50),

premium_amount DECIMAL(10,2),

policy_start DATE,

policy_end DATE

);



INSERT INTO insurance_policies VALUES

(301,'arjun',12000.50,'2023-01-01','2026-01-01'),

(302,'megha',8500.75,'2022-06-15','2025-06-15'),

(303,'vinod',15000.90,'2021-03-01','2024-03-01');
```

```sql
--QUESTION 8: Insurance Policy Aging
SELECT 
    policy_id,

    -- Uppercase holder name
    UPPER(holder_name) AS holder_name,

    -- Rounded premium
    ROUND(premium_amount, 0) AS premium_amount,

    policy_start,
    policy_end,

    -- 1. Policy duration in years
    TIMESTAMPDIFF(YEAR, policy_start, policy_end) AS duration_years,

    -- 2. Remaining days
    DATEDIFF(policy_end, CURRENT_DATE) AS remaining_days,

    -- 3. Policy Category
    CASE 
        WHEN policy_end < CURRENT_DATE THEN 'Expired'
        WHEN TIMESTAMPDIFF(YEAR, policy_start, policy_end) >= 3 THEN 'Long Term'
        WHEN TIMESTAMPDIFF(YEAR, policy_start, policy_end) = 2 THEN 'Mid Term'
        ELSE 'Short Term'
    END AS policy_status

FROM insurance_policies;
```
