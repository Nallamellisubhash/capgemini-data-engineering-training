Nested Case and When

```python
%sql
select * ,
case 
  when department ='Finance' then
  case
  when performance_rating ='A' and experience>=5 then  salary*0.25
  else salary*0.2 end
when department="HR" then
  case 
    when performance_rating ='B' and experience>=5 then salary*0.15
    else salary*0.1 end
else salary*0.08
end as bonus
from employee1

```

```python
%sql
select *, 
case 
when performance_rating='A' then 
  case 
  when salary>80000 and experience>5 then salary+salary*0.25
  when salary>=50000 and salary<80000 then salary+salary*0.15
  else salary+salary*0.2
  end
when performance_rating='B' then 
  case
  when experience>5 then salary+salary*0.12
  else salary+salary*0.1
  end
when performance_rating='C' then salary
end as hike 
from employee1
```

```python
%sql
select *,
case when salary>=70000 then 
case when performance_rating='A' and experience>=8 then "Top performer"
when experience<8 then "Mid performer"
else "good perfromer"
end
when salary>50000 and salary<70000 then 
case when performance_rating ='A' then "Rising_star"
else "Average performer"
end
else "low_performer"
end as performance
from employee1
```

```python
%sql
select *, 
case 
when salary>90000 then 
case 
when experience>10 then "35% tax bracket"
else "30% tax bracket"
end
when salary between 60000 and 90000 then 
case 
when experience>5 then "25% tax bracket"
else "20% tax bracket"
end
when salary<60000 then "15% tax bracket"
end as tax_bracketh 

from employee1
```
