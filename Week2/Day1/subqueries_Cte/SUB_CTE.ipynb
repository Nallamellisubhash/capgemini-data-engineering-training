```sql
--1. Sample Tables
-- Employees
CREATE TABLE Employees7 (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);
 --Departments
CREATE TABLE Departments7 (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50)
);
-- Orders
CREATE TABLE Orders7 (
    order_id INT PRIMARY KEY,
    emp_id INT,
    order_amount INT,
    order_date DATE
);
-- 2. Insert Sample Data
--Employees
INSERT INTO Employees7 VALUES
(1, 'Alice', 101, 60000, NULL),
(2, 'Bob', 102, 45000, 1),
(3, 'Charlie', 101, 70000, 1),
(4, 'David', 103, 40000, 2),
(5, 'Eve', 102, 50000, 2);
--Departments
INSERT INTO Departments7 VALUES
(101, 'HR'),
(102, 'IT'),
(103, 'Finance');
--Orders
INSERT INTO Orders7 VALUES
(1, 1, 5000, '2024-01-10'),
(2, 2, 3000, '2024-02-15'),
(3, 3, 7000, '2024-03-12'),
(4, 1, 2000, '2024-04-01'),
(5, 4, 1000, '2024-05-05');
```

```sql
--Find employees who earn more than average salary
select * from employees7 where salary>(select avg(salary) from employees7)
```

```sql
--Find employees who belong to the same department as 'Alice'
select * from employees7 where department_id=(select department_id from employees7 where name='Alice')
```

```sql
--Get employees whose salary is equal to the minimum salary
select * from employees7 where salary=(select min(salary) from employees7)
```

```sql
--Find employees who have placed at least one order

select * from employees7 where EXISTS(select orders7.emp_id from orders7 where orders7.emp_id = employees7.emp_id)
```

```sql
--Get employees whose salary is greater than ALL employees in department 102
select * from employees7 where salary > (select max(salary) from employees7 where department_id=102)
```

CTE

```sql
--Create a CTE to show employees with salary > 50,000 and fetch all records
with employee_cte as(select * from employees7 where salary>50000)
select * from employee_cte
```

```sql
--Find average salary per department using CTE
with department_avg as(
select department_id,avg(salary) from employees7 group by department_id)
select * from department_avg
```

```sql
--Use CTE to get employees and their department names
with employee_dept as(
select e.name,d.dept_name from employees7 e join departments7 d on e.department_id=d.dept_id)
select * from employee_dept
```

```sql
--Find total order amount per employee using CTE
with total_orders as(
select emp_id,sum(order_amount) as total_amount from orders7 group by emp_id)
select * from total_orders
```

```sql
--Find employees whose salary is above department average using CTE
with cte_emp(select e.* from employees7 e join (select department_id, avg(salary) as avg_salary from employees7 group by department_id) dept_avg on e.department_id = dept_avg.department_id where e.salary > dept_avg.avg_salary)
select * from cte_emp
```
